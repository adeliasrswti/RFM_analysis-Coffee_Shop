## Calculating RFM

```python
# Check the Maximum transaction date 
print(df1['transaction_date'].max())

# For calculationg recency, will use 2019-05-01
max_date = dt.datetime(2019, 5, 1)
print(max_date)
```

2019-05-01 will be used as the maximum date to calculate recency

```python
rfm = df1.groupby("customer_id").agg({"transaction_date" : lambda trx_date: (max_date - trx_date.max()).days,
                                     "transaction_id" : lambda trx_id: trx_id.nunique(),
                                     "line_item_amount" : lambda item_amount: item_amount.sum()})

# Name the columns as recency, frequency, and monetary
rfm.columns = ['recency', 'frequency', 'monetary']

print(rfm.head(15))
print(rfm.describe().T)
```

```python
# Rank the RFM Score
rfm['recency_score'] = pd.qcut(rfm['recency'].rank(method='first'), 5, labels=[5, 4, 3, 2, 1])
# rank method helps by assigning the first comer number to first label due to repetitive numbers
rfm['frequency_score'] = pd.qcut(rfm['frequency'], 5, labels=[1, 2, 3, 4, 5])
rfm['monetary_score'] = pd.qcut(rfm['monetary'], 5, labels=[1, 2, 3, 4, 5])

# Define rfm_score
rfm['RFM_score'] = (rfm['recency_score'].astype(str) + rfm['frequency_score'].astype(str))
# monetary score is necessary for observation, but it is not used in forming the RFM score
```

RFM Graph reference: https://medium.com/@denizcansuturan/rfm-analysis-3930a9a5238 


```python
# Segmentation Maps based on the RFM Graph
segm_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'   
}

# the R-F table is coded using regex
rfm['segment'] = rfm['RFM_score'].replace(segm_map, regex=True)
```

```python
rfm.head()
```

```python
segment_analysis = rfm.groupby('segment').agg({'recency': 'mean',
                                               'frequency': 'mean',
                                               'monetary': 'mean',
                                               'segment': 'count'})
segment_analysis.columns = ['r_mean', 'f_mean', 'm_mean', 'segm_count']
segment_analysis
```
